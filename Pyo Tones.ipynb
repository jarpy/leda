```python
import pyo
server = pyo.Server(audio="portaudio")
server.boot()
server.start()
tones = []
```

```python
tones.append(pyo.Sine(mul=0.1, freq=80).out())
```

```python
tones.append(pyo.Sine(mul=0.1, freq=120).out())
```

```python
tones.append(pyo.Sine(mul=0.1, freq=100).out())
```

```python
for tone in tones:
    tone.stop()
tones = []
```
